### 1. Sobre las variables predictoras. 
En este caso deberéis estandarizar estas variables. De nuevo, usad el método que prefiráis.


```python
df = pd.read_csv("../archivos/metro_2.csv", index_col=0)
df.head()
#Usamos el csv obtenido del archivo pp_regresion_lineal_05_B
```

```python
numericas = df.select_dtypes(include= np.number)
```

```python
numericas.drop(['traffic_volume', 'traffic_box'], axis= 1, inplace=True)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(numericas)
```

```python
transformacion = scaler.transform(numericas)
```

```python
df_transformacion = pd.DataFrame(transformacion, columns= numericas.columns).reset_index()
```

```python
df_transformacion
```

```python
df_mini= df[["weather_main", "traffic_box", "traffic_volume"]].reset_index()
df_mini
```

```python
df_mini.drop("index", axis=1, inplace=True)
```

```python
df_mini.reset_index()
```

```python
df_total= pd.concat([df_transformacion, df_mini], axis=1)
```

```python
df_total
```
