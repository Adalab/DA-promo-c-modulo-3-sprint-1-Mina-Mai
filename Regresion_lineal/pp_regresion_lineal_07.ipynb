```python
df= pd.read_csv('../archivos/metro_3.csv', index_col=0)
```

```python
df.head(2)
```

```python
lm = ols("traffic_box ~ temp + rain_1h + snow_1h + clouds_all + weather_main", data=df).fit()
```

```python
sm.stats.anova_lm(lm)
```

```python
lm.summary()
```
