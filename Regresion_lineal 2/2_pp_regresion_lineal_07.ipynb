```python
df= pd.read_csv('../archivos/metro_C.csv', index_col=0)
```

```python
df.head(2)
```

```python
lm = ols("traffic_box ~ temp_c + weekday + holiday_cat + snow_1h + time", data=df).fit()
```

```python
sm.stats.anova_lm(lm)
```

```python
lm.summary()
```

Viendo los resultados concluimos que todas nuestras variables predictoras tienen un valor menor a 0.05, excepto "snow_1h" junto a "holiday_cat", y nuestro R-squared aunque sigue siendo un poco bajo ha aumentado m谩s hasta un 8% en nuestro nuevo estudio.
